### Reformat the column names to make them easier to work with

```python
def make_list_removed_white_space_in_column_names(pandas_data_frame):
    list_column_names_removed_white_space = list()
    for column_name in eval(pandas_data_frame).columns:
        print(column_name)
        if column_name[0] == '\ufeff':   # because first column name in each frame includes this
            column_name = column_name.strip(u'\ufeff').replace('"', '')
        list_column_names_removed_white_space.append('_'.join(column_name.lower().split()))
    return list_column_names_removed_white_space
```

```python
for df_name in data_files:
    eval(df_name).columns = make_list_removed_white_space_in_column_names(df_name)
```

```python
def subset_by_indicator_name(indicator, pandas_data_frame):
    return pandas_data_frame[pandas_data_frame['indicator_name'] == indicator]
```
